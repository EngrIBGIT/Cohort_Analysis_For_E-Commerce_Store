This approach allows you to identify trends, anomalies, and areas where you may need to make adjustments to improve user retention and satisfaction.

## Create Cohort date:

*To help with segmenting according to dates*

- Since we are carrying out a Time-based cohort analysis, the cohorts will be grouped according to the dates customers made their first purchase

**and**

- Retention rate will be calculated by analysis the months they stayed active after their first purchase

```python
#Create Invoice month as new feature
def get_month(x):
    return dt.datetime(x.year, x.month, 1)
```

```python
data["InvoiceDate"]  = data["InvoiceDate"].apply(get_month)
data.tail()
```

```python
def get_cohort_date(data):
    """ this function takes in the dataframe
        and returns the cohort date

        variables:
        data = dataframe
        cohort date = the first date they made a purchase
        """

    # assign the minimum date(first purchase date) to all unique customerID, i.e the first day they made a purchase
    data["cohort date"] = data.groupby("CustomerID")["InvoiceDate"].transform("min")

    return data["cohort date"]
```

```python
#apply the function created to our dataframe and extract the 2 newly created column

data["cohort date"] = get_cohort_date(data)
```

```python
data
```

## Create Cohort index

A cohort index is a numerical representation that measures the time interval in months since a particular group (cohort) made their initial purchase.

For example, an index of 4 indicates that this cohort made their first purchase four(4) months ago.

This will be calculated by subtracing the time the customer made first purchase (cohort date) from the recent purchase time (InvoiceDate)

```python
#create a function that extracts the year and month from the first and last cohort date

def get_year_and_month(data, col):
    """
    This function takes in the dataframe and column,
    and returns the month and year component for that column

    Variables:
    data = dataframe
    col = column
    month = month component
    year = year component"""

    month = data[col].dt.month
    year = data[col].dt.year
    return month, year
```

```python
# apply the fucntion on cohort first date column
first_month, first_year = get_year_and_month(data,"cohort date")
```

```python
first_month
```

```python
first_year
```

```python
# apply the fucntion on cohort latest date column
latest_month, latest_year = get_year_and_month(data,"InvoiceDate")
```

```python
latest_month
```

```python
latest_year
```

**CREATE COHORT INDEX**

```python
# write a function to create cohort index
def create_cohort_index(first_month, first_year, latest_month,latest_year):
    """
    This code takes in the first_month, first_year, latest_month and latest_year
    and returns the calculated period(in months) the customer has been active

    variables:
    first_month: first month the customer made purchase
    first_year: first year the customer made purchase

    latest_month: recent month the customer made purchase
    latest_year: recent year the customer made purchase

    index: The duration between first and latest purchase (in months)"""

    year_diff = latest_year - first_year
    month_diff = latest_month - first_month
    index = year_diff*12 + month_diff +1 # +1 is added because of customers who have been active for just 1 month
    return index
```

```python
data["cohort_index"] = create_cohort_index(first_month, first_year, latest_month, latest_year)
```

# **Segmentation Analysis**

Segmentation analysis will help us understand the purchasing behavior of customers by grouping them based on their purchasing patterns.

We will be using the `RFM (Recency, Frequency, Monetary) Technique.` which uses three(3) key metrics (Recency, Frequency, Monetary)

**RFM Analysis**

RFM (Recency, Frequency, Monetary) analysis is a common technique used to segment customers based on their purchasing behavior.

The three (3) key metrics explains:
- Recency (how recently a customer made a purchase),
- Frequency (how often a customer makes purchases), and
- Monetary Value (how much money a customer spends).


```python
data.head()
```

**Calculate  the Recency, Frequency, and Monetary (RFM) Values:**


- We will find the most recent invoice date (max) in the "InvoiceDate" column of the data DataFrame.
-Then add one day (dt.timedelta(days=1)) to this date to account for invoices potentially occurring on the current day.

This ensures all invoices are considered for RFM calculations.

```python
# Calculate Recency, Frequency, and Monetary values
the_date = data["InvoiceDate"].max() + dt.timedelta(days=1)

rfm = data.groupby("CustomerID").agg({
    "InvoiceDate": lambda x: (the_date - x.max()).days,
    "InvoiceNo": "nunique",
    "UnitPrice": "sum"
})

# Renaming columns:
rfm.rename(columns={
    "InvoiceDate": "Recency",
    "InvoiceNo": "Frequency",
    "UnitPrice": "MonetaryValue"
}, inplace=True)
```

```python
rfm.head()
```

**Creating RFM Segments:**

Using the pd.qcut function in pandas to create quartiles (i.e groups) based on the distribution of Recency, Frequency, and Monetary (RFM) Value.

To create a new column named `R_Quartile` within the rfm DataFrame.
which will use pd.qcut to divide the "Recency" values into four (4) equal quartiles (groups).

The labels argument assigns labels (weights) in descending order (4 for the highest recency, 1 for the lowest) to each quartile.

This indicates how recent a customer's purchase was within their group.

Similar logic is applied to create `F_Quartile` (Frequency) and `M_Quartile` (Monetary Value) with labels (1 to 4) representing their respective ranks within each quartile.

```python
# Create RFM segments
rfm["R_Quartile"] = pd.qcut(rfm["Recency"], 4, labels=[4, 3, 2, 1])
rfm["F_Quartile"] = pd.qcut(rfm["Frequency"].rank(method = 'first'), 4, labels=[1, 2, 3, 4])
rfm["M_Quartile"] = pd.qcut(rfm["MonetaryValue"], 4, labels=[1, 2, 3, 4])
```

```python
rfm
```

C**ombining the RFM Segments:**

We combine the RFM quartiles (representing ranks within each metric) into a single string column named `RFM_Segment`.

It concatenates the string representations of each quartile (e.g., "413" for a customer in the highest recency quartile, second frequency quartile, and third monetary value quartile).

```python
# Combine RFM segments into a single string for easier identification
rfm["RFM_Segment"] = rfm["R_Quartile"].astype(str) + rfm["F_Quartile"].astype(str) + rfm["M_Quartile"].astype(str)
```

```python
rfm
```

**Calculating RFM Score:**

We calculate an `RFM score` for each customer by summing their RFM quartile values (weights) in a new column named "`RFM_Score`".

A higher RFM score indicates a customer with more recent purchases, higher purchase frequency, and greater monetary value.

```python
# Calculate RFM Score
rfm["RFM_Score"] = rfm[["R_Quartile", "F_Quartile", "M_Quartile"]].sum(axis=1)
```

```python
# Segment customers based on RFM Score
rfm["RFM_Level"] = pd.qcut(rfm["RFM_Score"], 3, labels=["Low", "Medium", "High"])

rfm.head()
```

# **Improved Segmentation Analysis with Advanced Clustering Techniques and Geospatial Analysis**



We will investigate the following clustering methods  K-Means, DBSCAN, and Hierarchical Clustering for improved segmentation analysis in order to divide up our clientele according to their purchase patterns.

To learn more about client behavior, we will also use geospatial analysis and customer journey mapping.

## Data Preparation

We prepare the data before clustering methods are applied. This entails managing any missing values and scaling the features.

```python
# Select features for clustering
X = rfm[["Recency", "Frequency", "MonetaryValue"]]

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## K-Means Clustering
K-Means is a centroid-based clustering algorithm that partitions the data into k clusters, where each data point belongs to the cluster with the nearest mean.

Process:
Determine the optimal number of clusters using the Elbow Method.

Fit the K-Means model.

Assign clusters to customers.

Visualize the clusters.

```python
# Determine the optimal number of clusters using the Elbow Method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
# Plot the Elbow Method
plt.figure(figsize=(20, 10))
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method for Optimal k", fontsize=20, fontweight='bold')
plt.xlabel("Number of Clusters (k)", fontsize=12, fontweight='bold')
plt.ylabel("Inertia", fontsize=12, fontweight='bold')
plt.show()
```

```python
# Fit K-Means with the optimal number of clusters (k=3)
kmeans = KMeans(n_clusters=3, random_state=42)
rfm["KMeans_Cluster"] = kmeans.fit_predict(X_scaled)
```

```python
# Visualize the clusters
plt.figure(figsize=(20, 10))
sns.scatterplot(x="Frequency", y="MonetaryValue", hue="KMeans_Cluster", data=rfm, palette="viridis")
plt.xlabel("Frequency", fontsize=12, fontweight='bold')
plt.ylabel("MonetaryValue", fontsize=12, fontweight='bold')
plt.title("K-Means Clustering", fontsize=20, fontweight='bold')
plt.show()
```

## Observations:
The Elbow Method helps identify the optimal number of clusters (e.g., k=3).

Customers are grouped into distinct clusters based on their purchasing behavior.

## Recommendations:
High-Value Customers: Focus on retaining these customers with loyalty programs.

Medium-Value Customers: Encourage repeat purchases through targeted marketing.

Low-Value Customers: Re-engage with personalized offers.

## DBSCAN Clustering
DBSCAN (Density-Based Spatial Clustering of Applications with Noise) is a density-based clustering algorithm that groups together points that are closely packed and marks outliers as noise.

Process:

Fit the DBSCAN model.

Assign clusters to customers.

Visualize the clusters.

```python
# Fit DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
rfm["DBSCAN_Cluster"] = dbscan.fit_predict(X_scaled)
```

```python
# Visualize the clusters
plt.figure(figsize=(20, 10))
sns.scatterplot(x="Frequency", y="MonetaryValue", hue="DBSCAN_Cluster", data=rfm, palette="viridis")
plt.title("DBSCAN Clustering", fontsize=20, fontweight='bold')
plt.xlabel("Frequency", fontsize=12, fontweight='bold')
plt.ylabel("MonetaryValue", fontsize=12, fontweight='bold')
plt.show()
```

Observations:

DBSCAN identifies dense regions and outliers (noise).

Useful for detecting anomalies or unusual customer behavior.



Recommendations:

Core Customers: Focus on retaining these high-density clusters.

Outliers: Investigate unusual behavior (e.g., one-time purchasers or potential fraud).
